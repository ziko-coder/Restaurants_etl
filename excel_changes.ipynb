## Separate Categories into different columns
    If a restaurant doesn't have 3 categories, the appropriate cell is filled
    with 'None' value 


```python
categories = df.Categories
separated = []
for category in categories:
    els = category.split(',')
    separated.append(els)
    
cat_1 = []
cat_2 = []
cat_3 = []
cat_4 = []

for el in separated:
    l = len(el)
    if l >= 1:
        cat_1.append(el[0])
    else:
        cat_1.append('None')

    if l >= 2:
        cat_2.append(el[1])
    else:
        cat_2.append('None')

    if l >= 3:
        cat_3.append(el[2])
    else:
        cat_3.append('None')

    if l >= 4:
        cat_4.append(el[3])
    else:
        cat_4.append('None')
```

```python
df['Category_1'] = cat_2
df['Category_2'] = cat_3
df['Category_3'] = cat_4
df = df.drop('Categories', axis=1)
```

```python
len(cat_3), len(cat_4), len(cat_2), len(cat_1)
```

## Separate Amenities into different columns
    If a restaurant doesn't have 7 amenities, the appropriate cell is filled
    with 'None' value 

```python
amenities = df.Amenities
separated_amenities = []
for amenity in amenities:
    if isinstance(amenity, str):
        els = amenity.split(',')
        separated_amenities.append(els)
    else:
        separated_amenities.append(['None']) 
```

```python
set(len(el) for el in  separated_amenities)
```

```python
am_1 = []
am_2 = []
am_3 = []
am_4 = []
am_5 = []
am_6 = []
am_7 = []

for el in separated_amenities:
    l = len(el)
    if l >= 1:
        am_1.append(el[0])
    else:
        am_1.append('None')

    if l >= 2:
        am_2.append(el[1])
    else:
        am_2.append('None')

    if l >= 3:
        am_3.append(el[2])
    else:
        am_3.append('None')

    if l >= 4:
        am_4.append(el[3])
    else:
        am_4.append('None')
    if l >= 5:
        am_5.append(el[4])
    else:
        am_5.append('None')
    if l >= 6:
        am_6.append(el[5])
    else:
        am_6.append('None')
    if l >= 7:
        am_7.append(el[6])
    else:
        am_7.append('None')
```

```python
for i in range(7):
    print(len(eval(f'am_{i+1}')))
```

```python
for i in range(7):
    col_name = f'Amenity_{i+1}'
    df[col_name] = eval(f'am_{i+1}')
    
df = df.drop('Amenities', axis=1)
```

## Modifed the "Years in Business" Column

```python
df['Years in Business'] = df['Years in Business'].str.slice(stop=2)

years_modified = []
for el in df['Years in Business']:
    if el == 'NO':
        years_modified.append(0)
    else:
        years_modified.append(el)
```

## Converted 'Price Range' column to Strings

```python
df['Price Range'].value_counts().keys()
```

```python
for i in range(len(df)): 
    if df['Price Range'][i] == '$':
        df['Price Range'][i] = 'cheap'
    elif df['Price Range'][i] == '$$':
        df['Price Range'][i] = 'moderate'
    elif df['Price Range'][i] == '$$$':
        df['Price Range'][i] = 'highly_priced'  
    elif df['Price Range'][i] == '$$$$':
        df['Price Range'][i] = 'expensive'    
```

## Converted 'Stars' column to numerical values

```python
df['Stars'].value_counts().keys()
```

```python
for i in range(len(df)): 
    if df['Stars'][i] == 'one':
        df['Stars'][i] = 1
    elif df['Stars'][i] == 'two':
        df['Stars'][i] = 2
    elif df['Stars'][i] == 'three':
        df['Stars'][i] = 3  
    elif df['Stars'][i] == 'four':
        df['Stars'][i] = 4
    elif df['Stars'][i] == 'five':
        df['Stars'][i] = 5
```
